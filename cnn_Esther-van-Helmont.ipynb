## 2. Data & preprocessing – “plaatjes klaarmaken als oefenblaadjes”

Wat we deden met de beelden:

- **Resize naar 256 → Crop naar 224**  
  Vergelijk: je knipt alle foto’s netjes uit tot een standaard formaat, zodat het model niet eerst puzzelt met allerlei maten.

- **Train‑augmentatie: `RandomHorizontalFlip`**  
  Alsof je dezelfde foto soms in de spiegel bekijkt: helpt het model om niet alleen één exacte hoek te herkennen.

- **Normalize met `mean` en `std`**  
  We trekken gemiddeldheid eraf en delen door de spreiding.  
  Metafoor: je zet alle cijfers eerst om naar “rapportcijfers tussen 1 en 10”, ongeacht de originele schaal, zodat leren makkelijker wordt.

Effect:  
- Trainset is iets “moeilijker” (door flip).  
- Valid/test zijn schoner (alleen center crop + normalisatie).

---

## 3. Modelarchitectuur – “hoeveel stappen denkt de leerling na”

### Eerst: conv‑blokken

- Meerdere lagen `Conv → BatchNorm → ReLU → MaxPool`.
- Vergelijk: de leerling kijkt eerst naar:
  - randen,
  - dan vormen,
  - dan patronen (ramen, torens, bogen).

Elke laag pakt steeds complexere kenmerken uit de afbeelding.

### Belangrijke wijziging: **Global Average Pooling**

- **Oude versie:**
  - Na de convoluties hadden we een groot rooster (256 × 14 × 14 → 50 176 waarden).
  - Die stopten we direct in een grote `Linear` laag (50176 → 512).  
  Metafoor: een leerling die 50 000 losse feitjes uit het hoofd moet stampen. Heel zwaar, moeilijk te optimaliseren.

- **Nieuwe versie:**
  - We pakken per kanaal het **gemiddelde** over ruimte (`AdaptiveAvgPool2d(1,1)`), dus van 256×14×14 → 256×1×1.
  - Daarna pas `Flatten → Linear(256 → 512 → 50)`.  
  Metafoor: in plaats van elk los detail te onthouden, maakt de leerling per thema een samenvatting (“overal zie ik water / rotsen / torens”).

Effect:

- Veel minder parameters → makkelijker trainen.
- Minder snel vastlopen in slechte oplossingen.
- Dit was een grote reden waarom we van ~18 % naar ~59 % konden gaan.

### Dropout = “af en toe stukje weggooien bij oefenen”

- `dropout = 0.3`: bij het leren zetten we 30 % van de neuronen tijdelijk op 0.
- Metafoor: je leert niet altijd met dezelfde samenvatting; soms mis je een stukje papier en moet je de rest begrijpen.
- Doel: model leert **algemene** patronen, niet één specifieke set “trucjes” → minder overfitting.

---

## 4. Training & loss – “hoe meten we hoe slecht de leerling het doet”

### CrossEntropyLoss

- Voor 50 klassen:
  - Willekeurig gokken ≈ 2 % accuracy.
  - De bijbehorende loss ≈ `-log(1/50) ≈ 3.91`.

- Als we loss zien als “hoeveel fouten in het proefwerk”:
  - Loss 3.9 → echt slecht, bijna random.
  - Loss 1.7 → veel beter; het model geeft hogere kans aan de juiste klasse.

We hebben gezien:

- **Slechte runs:** loss bleef rond 3.2–3.8 → bijna random → accuracy 7–20 %.
- **Goede run:** test loss **1.69** → veel beter → **59 % accuracy**.

## 5. Optimizer & hyperparameters – “hoe snel en netjes de leerling bijstuurt”

### Learning rate (`learning_rate = 0.0015`)

- Metafoor: grootte van de stap waarmee de leerling zijn aanpak verandert na een fout.
  - Te laag (heel kleine stapjes) → je komt haast niet vooruit.
  - Te hoog (reuzensprongen) → je stuitert langs het goede antwoord heen.

- Eerst probeerden we lagere lr’s (0.0005), die bleken **te voorzichtig**: verlies zakte amper.
- 0.0015 gaf een flinke daling van de trainingsloss (naar ~1.3).

### Adam optimizer

- Adam ≈ slimme leerling die:
  - bijhoudt wat hij in het verleden fout deed,
  - sneller corrigeert waar vaak fouten zitten,
  - rustiger wordt op plekken die al goed zijn.

- Voor dit probleem was Adam duidelijk effectiever dan onze eerdere mixen met andere instellingen.

### Weight decay (`weight_decay = 1e-4`)

- Metafoor: straf op extreem hoge gewichten = “niet te zelfverzekerd zijn”.
- Zonder weight decay kan het model sommige neuronen gigantisch belangrijk maken → kan overfitten (te veel “van buiten leren”).

- Kleine waarde 1e‑4 gaf voldoende rem, maar hield het leren toch vlot.

### Scheduler `ReduceLROnPlateau`

- Kijkt naar **validatieloss**; als die een tijdje niet meer beter wordt:
  - halveert hij de lr.

- Metafoor: als je merkt dat grote stappen niet meer helpen, ga je kleinere stapjes zetten om te verfijnen.

- Je zag in de lr‑grafiek:
  - eerst vlak op 0.0015,
  - dan sprong naar ~0.00075,
  - dan naar ~0.000375.

- Hierdoor konden we eerst snel omlaag, daarna rustig “bijschaven”.
